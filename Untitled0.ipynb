```python
import numpy as np
import matplotlib.pyplot as plt

import pandas as pd
url='https://raw.githubusercontent.com/nastya12132y3782y3487/MMVSULAB1/main/Omsk.csv'
df=pd.read_csv(url)
print(df)
years=np.array(df['Years'])
population=np.array(df['Population'])
plt.plot(years,population)
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Population of Omsk since 1926')
plt.show()
```
